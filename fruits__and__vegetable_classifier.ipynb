```python
DATADIR = "/content/images"
CATEGORIES = ['apple', 'banana', 'beetroot', 'bell pepper', 'cabbage', 'capsicum', 'carrot', 'cauliflower',
              'chilli pepper', 'corn', 'cucumber', 'eggplant', 'garlic', 'ginger', 'grapes', 'jalepeno',
              'kiwi', 'lemon', 'lettuce', 'mango', 'onion', 'orange', 'paprika', 'pear', 'peas',
              'pineapple', 'pomegranate', 'potato', 'radish', 'soy beans', 'spinach', 'sweetcorn',
              'sweetpotato', 'tomato', 'turnip', 'watermelon']
for category in CATEGORIES:
    class_num = CATEGORIES.index(category)
    path = os.path.join(DATADIR, category)
    for img in os.listdir(path):
      img_array = imread(os.path.join(path, img))
      #print(img_array.shape)
      #plt.imshow(img_array)
      img_resized = resize(img_array,(150,150,3))
      flat_data.append(img_resized.flatten())
      images.append(img_resized)
      target.append(class_num)

flat_data = n.array(flat_data)
target = n.array(target)
images = n.array(images)
```

```python
num_samples_to_keep = len(target)
flat_data = flat_data[:num_samples_to_keep]
```

```python
unique,count = n.unique(target,return_counts=True)



num_samples_to_keep = len(target)
flat_data = flat_data[:num_samples_to_keep]



```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data, target,test_size=0.3,random_state=109)
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'kernel':['linear']},
              {'C':[1,10,100,1000],'kernel':['linear']},
]
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```
